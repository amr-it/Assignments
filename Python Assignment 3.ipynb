1. Why are functions advantageous to have in your programs?

Functions in Python. You use functions in programming to bundle a set of instructions that you want to use repeatedly or that, because of their complexity, are better self-contained in a sub-program and called when needed. That means that a function is a piece of code written to carry out a specified task.

2. When does the code in a function run: when it's specified or when it's called?

Function runs once it's referenced. Hence, once a function is called, then, code inside it runs.

3. What statement creates a function?

In Python, you define a function with the def keyword, then write the function identifier (name) followed by parentheses and a colon. The next thing you have to do is make sure you indent with a tab or 4 spaces, and then specify what you want the function to do for you.

4. What is the difference between a function and a function call?

Function is a set of instructions that can be used repeatedly, whereas, to "call" means to make a reference in your code to a function that is written elsewhere.

5. How many global scopes are there in a Python program? How many local scopes?

There is 1 global scope in Python, while, local scopes are created every time you call a function.

6. What happens to variables in a local scope when the function call returns?

Parameters and names that you assign inside a function exist only within the function or local scope associated with the function call. When the function returns, the local scope is destroyed and the names are forgotten.
